```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## EXP measure the acceleration of gravity
data source: Tracker

```python
df1 = pd.read_csv('ball.csv')
df1 = df1.drop('Unnamed: 0', axis=1)
print(df1)
```

```python
t2 = df1['t']
y2 = df1['y']
plt.scatter(t2, y2, color='lightseagreen', label='original curve')
def y_t(t, g, v0):
    return (1/2)*g*t**2 + v0*t

popt, pcov = curve_fit(y_t, t2, y2)
g, v0 = popt
gvals = y_t(t2, g, v0)
plt.plot(t2, gvals, color='lightcoral', label='fitting curve; fit: g=%5.3f, v0=%5.3f ' % tuple(popt))

plt.xlabel('Time t/s')
plt.xticks(list(np.arange(0, 0.55, 0.05)))
plt.ylabel('Height y/m')
plt.title('Free fall of the small ball', y=-.25)
plt.legend()
```
