```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df1 = pd.read_csv('B12-Data/calibration.csv')
cal_func = np.polyfit(df1['T'], df1['U'], deg=1)

T_dns = np.linspace(0, 100, 500)
U_reg = np.polyval(cal_func, T_dns)

r = np.corrcoef(df1['T'], df1['U'])[0,1]

A1, A2 = tuple(cal_func)
plt.scatter(df1['T'], df1['U'], color='lightcoral', label='Empirical data')
plt.plot(T_dns, U_reg, color='lightseagreen', 
                          label='Fitting curve. \ny = {:.4f}x + {:.4f}\nr = {:.8f}'
                          .format(A1, A2, r))
plt.legend()

plt.xlabel('Temperature t/$^oC$')
plt.ylabel('Voltage of the sensor/$\Omega$')
plt.title('fig. 1 The calibration curve of the AD590 sensor.', y=-0.25)
plt.savefig(r'B12-Output\fig.1.png', bbox_inches = 'tight', dpi=300)
```
