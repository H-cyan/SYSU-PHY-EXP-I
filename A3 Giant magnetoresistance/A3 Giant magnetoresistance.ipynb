```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import Rbf

u = 4*np.pi*1e-7
n = 2.4*1e4
```

## EXP 1

```python
df1 = pd.read_csv('A3 data\EXP1.csv')
df1['B'] = u*n*df1['I']*10
df1 = df1[['I', 'B', 'U1', 'U2', 'U3', 'U4']]
df1
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(df1['B'], df1['U1'], 'o-', color='#60c5ba', label='Magnetic Flux Density decreasing')
plt.plot(-df1['B'], df1['U2'], 'o-', color='#60c5ba')

plt.annotate(text='', xy=(5, 150), xytext=(10, 200), arrowprops=dict(arrowstyle='->'
                                ,color='#60c5ba',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=0.3"))
plt.annotate(text='', xy=(-10, 200), xytext=(-5, 150), arrowprops=dict(arrowstyle='->'
                                ,color='#60c5ba',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=0.3"))

plt.plot(df1['B'], df1['U3'], '*-', color='#ef5285', label='Magnetic Flux Density increasing')
plt.plot(-df1['B'], df1['U4'], '*-', color='#ef5285')

plt.annotate(text='', xy=(20, 200), xytext=(15, 150), arrowprops=dict(arrowstyle='->'
                                ,color='#ef5285',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=-0.3"))
plt.annotate(text='', xy=(-15, 150), xytext=(-20, 200), arrowprops=dict(arrowstyle='->'
                                ,color='#ef5285',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=-0.3"))

plt.xlabel('Magnetic Flux Density B/Gs')
plt.ylabel('Output Voltage U/mV')
plt.title('Fig. 1 Magnetoelectric conversion of the GMR sensor', y=-0.3)
plt.grid()
plt.legend(loc='center left', bbox_to_anchor=(0.1, -0.17), ncol=2)
plt.savefig(r'A3 output\fig. 1.png', bbox_inches = 'tight', dpi=300)
```

## EXP3

```python
df3 = pd.read_csv('A3 data/EXP3.csv')
df3.set_index('u/d', drop=True, inplace=True)
df3['B_cls'] = u*n*df3['I_cls']*10
df3['B_opn'] = u*n*df3['I_opn']*10
df3 = df3[['I_cls', 'B_cls', 'I_opn', 'B_opn']]
df3
```

```python
i1 = np.linspace(df3.loc['d', 'B_cls'], 15.079645, 100)
i2 = np.linspace(df3.loc['d', 'B_opn'], df3.loc['d', 'B_cls'], 100)
i3 = np.linspace(-15.079645, df3.loc['d', 'B_opn'], 100)

U_h = np.array([1]*100)
U_l = np.array([-1]*100)

dn = np.array([np.concatenate((i3, i2, i1), axis=0), np.concatenate((U_h, U_l, U_h), axis=0)])
```

```python
I1 = np.linspace(-15.079645, df3.loc['u', 'B_cls'], 100)
I2 = np.linspace(df3.loc['u', 'B_cls'], df3.loc['u', 'B_opn'], 100)
I3 = np.linspace(df3.loc['u', 'B_opn'], 15.079645, 100)



U_h = np.array([1]*100)
U_l = np.array([-1]*100)

up = np.array([np.concatenate((I1, I2, I3), axis=0), np.concatenate((U_h, U_l, U_h), axis=0)])
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(up[0], up[1], color='#60c5ba', label='Magnetic Flux Density decreasing')
plt.plot(dn[0], dn[1], color='#ef5285', label='Magnetic Flux Density increasing')




plt.annotate(text='', xy=(-3, 0.75), xytext=(-0.5, 0.25), arrowprops=dict(arrowstyle='->'
                                ,color='#60c5ba',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=0.3"))
plt.annotate(text='', xy=(0, 0.25), xytext=(2.5, 0.75), arrowprops=dict(arrowstyle='->'
                                ,color='#60c5ba',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=0.3"))

plt.annotate(text='', xy=(-7, 0.25), xytext=(-10, 0.75), arrowprops=dict(arrowstyle='->'
                                ,color='#ef5285',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=-0.3"))
plt.annotate(text='', xy=(10, 0.75), xytext=(7, 0.25), arrowprops=dict(arrowstyle='->'
                                ,color='#ef5285',
                                shrinkA = 5,
                                shrinkB = 5,
                                connectionstyle="arc3,rad=-0.3"))

plt.xlabel('Magnetic Flux Density B/Gs')
plt.ylabel('Output Voltage U/mV')
plt.title('Fig. 3 Magnetoelectric conversion of the GMR switch', y=-0.3)
plt.grid()
plt.legend(loc='center left', bbox_to_anchor=(0.1, -0.17), ncol=2)
plt.savefig(r'A3 output\fig. 3.png', bbox_inches = 'tight', dpi=300)
```
