```python
def stocks(angle, p):
    px, py, p45, pr = p
    s0 = px + py
    s1 = px - py
    s2 = 2*p45 - (px + py)
    s3 = (px + py) - 2*pr
    phi, chi = angle
    print("Experimental:")
    print(s0, s1, s2, s3)
    print("normalized")
    print(s0/s0, s1/s0, s2/s0, s3/s0)
    print("Theoretical:")
    s0 = (s1**2+ s2**2+ s3**2)**(1/2)
    s1 = s0*np.cos(2*chi)*np.cos(2*phi)
    s2 = s0*np.cos(2*chi)*np.sin(2*phi)
    s3 = s0*np.sin(2*chi)
    print(s0, s1, s2, s3)
    print("normalized")
    print(s0/s0, s1/s0, s2/s0, s3/s0)
```

```python
stocks((np.pi/4, 0), (0.315, 0.305, 0.615, 0.321))
```

```python
stocks((13*np.pi/36, 0), (0.105, 0.508, 0.535, 0.290))
```

```python
stocks((0, -np.pi/4), (0.307, 0.294, 0.312, 0.602))
```
