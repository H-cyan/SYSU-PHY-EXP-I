```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'B4 data\EXP3\Damp4V Drive0V loc4.txt', sep='\t')
df = df.dropna()
```

```python
loc = 4*7 +1
df = df.iloc[:, loc:loc+2]
df = df.dropna()

df.iloc[:,0] = df.iloc[:,0] - df.iloc[0,0]
df.iloc[:,1] = df.iloc[:,1] - df.iloc[0,1]
df.reset_index(drop=True, inplace=True)
df.head(10)

sqrphi_origin = (np.array(df.iloc[:, 0]))**2
sqromega_origin = (np.array(df.iloc[:, 1]))**2
```

```python
plt.plot(sqrphi_origin, sqromega_origin, linewidth=0.5, color='lightseagreen')

plt.xlabel('phi^2')
plt.ylabel('omega^2')
plt.title('fig. 3.1 Damped vibration of the Bohr pendulum.\nDamp=4V Drive=0V', y=-0.3)
```
